## ONLY DONE ONCE CHECK DAY FOLDERS 'MMDD' IF .JSON FILES EXIST

```python
day = '0115'
direct = '/data/ana/Calibration/POCAM/data/freezein/device/'
file_path = os.path.join(direct,day)
files = [file for file in os.listdir(file_path) if file.endswith(".log")]
files.sort()
print(files)
```

```python
for file in files:
    timestamps = []
    dic = {}
    name_splits = file.split('_')
    device_surfix = name_splits[2]+'_'+name_splits[3]
    
    with open(os.path.join(file_path,file), "r") as f:
        for line in f:
            ts_str = line[:23]  # "2026-01-06 19:18:29,232"
            if ts_str=='\n':
                continue
            timestamps.append(ts_str)

    run_start = timestamps[0]
    run_end   = timestamps[-1]
    first_log = timestamps[2]
    end_log   = timestamps[-11]

    file_name = 'run_start_end_emitter_start_end_'+device_surfix+'.json'
    dic['run_start'] = run_start
    dic['run_end']   = run_end
    dic['first_log'] = first_log
    dic['end_log']   = end_log

    with open(os.path.join(file_path,file_name), 'w') as file:
        json.dump(dic, file, indent=4)
```
